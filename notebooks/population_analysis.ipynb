# Population Data Analysis

This notebook analyzes population data from the World Bank dataset to create visualizations using different plotting techniques.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample demographic data
np.random.seed(42)
n_samples = 1000

# Generate age data with a normal distribution (mean age 35, std 15)
ages = np.random.normal(35, 15, n_samples)
ages = np.clip(ages, 0, 90).astype(int)  # Clip ages between 0 and 90

# Generate gender data
genders = np.random.choice(['Male', 'Female'], n_samples, p=[0.48, 0.52])

# Create a DataFrame
df = pd.DataFrame({
    'Age': ages,
    'Gender': genders
})
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
#gender Distribution (Bar Plot)
plt.subplot(1, 2, 2)
gender_counts = df['Gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Gender Distribution in Population')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Box Plot: Population Distribution within Regions
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Region', y='Population')
plt.title('Population Distribution by Region')
plt.xticks(rotation=45)
plt.ylabel('Population')
plt.show()
```

```python
# Pie Chart: Population Distribution by Region
plt.figure(figsize=(10, 8))
pop_by_region = df.groupby('Region')['Population'].sum()
plt.pie(pop_by_region, labels=pop_by_region.index, autopct='%1.1f%%')
plt.title('Population Distribution by Region')
plt.show()
```

```python
# Load the data
# You can either download the CSV from World Bank or use this sample data
data = {
    'Country': ['China', 'India', 'USA', 'Indonesia', 'Pakistan', 'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico'],
    'Population': [1412600000, 1380004385, 331002651, 273523615, 220892340, 212559417, 206139589, 164689383, 144104080, 128932753],
    'Region': ['Asia', 'Asia', 'North America', 'Asia', 'Asia', 'South America', 'Africa', 'Asia', 'Europe', 'North America']
}

df = pd.DataFrame(data)
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Country', y='Population')
plt.xticks(rotation=45)
plt.title('Population by Country (2021)')
plt.ylabel('Population')
plt.xlabel('Country')
plt.show()
```
