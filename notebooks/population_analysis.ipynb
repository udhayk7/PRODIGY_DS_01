# Population Data Analysis

This notebook analyzes population data from the World Bank dataset to create visualizations using different plotting techniques.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visualizations
plt.style.use('default')
sns.set_theme(style='whitegrid')
```

```python
# Box Plot: Population Distribution within Regions
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Region', y='Population')
plt.title('Population Distribution by Region')
plt.xticks(rotation=45)
plt.ylabel('Population')
plt.show()
```

```python
# Pie Chart: Population Distribution by Region
plt.figure(figsize=(10, 8))
pop_by_region = df.groupby('Region')['Population'].sum()
plt.pie(pop_by_region, labels=pop_by_region.index, autopct='%1.1f%%')
plt.title('Population Distribution by Region')
plt.show()
```

```python
# Scatter Plot: Population vs GDP
df['GDP_Billions'] = [14722, 2622, 20937, 1058, 263, 1445, 440, 324, 1483, 1076]  # Sample GDP data
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Population', y='GDP_Billions')
plt.title('Population vs GDP')
plt.xlabel('Population')
plt.ylabel('GDP (Billions USD)')
# Add country labels to points
for i, txt in enumerate(df['Country']):
    plt.annotate(txt, (df['Population'].iloc[i], df['GDP_Billions'].iloc[i]))
plt.show()
```

```python
# Load the data
# You can either download the CSV from World Bank or use this sample data
data = {
    'Country': ['China', 'India', 'USA', 'Indonesia', 'Pakistan', 'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico'],
    'Population': [1412600000, 1380004385, 331002651, 273523615, 220892340, 212559417, 206139589, 164689383, 144104080, 128932753],
    'Region': ['Asia', 'Asia', 'North America', 'Asia', 'Asia', 'South America', 'Africa', 'Asia', 'Europe', 'North America']
}

df = pd.DataFrame(data)
df.head()
```

## Bar Chart: Population by Country

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Country', y='Population')
plt.xticks(rotation=45)
plt.title('Population by Country (2021)')
plt.ylabel('Population')
plt.xlabel('Country')
plt.show()
```

## Distribution of Population (Histogram)

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Population', bins=15)
plt.title('Distribution of Population Across Countries')
plt.xlabel('Population')
plt.ylabel('Count')
plt.show()
```

## Countries by Region (Bar Chart)

```python
region_counts = df['Region'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=region_counts.index, y=region_counts.values)
plt.title('Number of Countries by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

## Average Population by Region

```python
avg_pop_by_region = df.groupby('Region')['Population'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_pop_by_region.index, y=avg_pop_by_region.values)
plt.title('Average Population by Region')
plt.xlabel('Region')
plt.ylabel('Average Population')
plt.xticks(rotation=45)
plt.show()
```
